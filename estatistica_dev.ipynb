```python
import pandas as pd;

pd.__version__

data_medidas = {
    'idade': [15, 18, 25, 25, 40, 55, 58, 60, 80],
    'altura': [160,165,168,172,174,174,174,190, 200],
}

df_medidas = pd.DataFrame.from_dict(data_medidas);


df_medidas
```

## Medidas Resumo

### Medidas de posição

```python
# Media

df_medidas["idade"].mean()
```

```python
# Mediana

df_medidas["idade"].median()
```

```python
# Moda

df_medidas["idade"].mode()
```

```python
# Media

df_medidas["altura"].mean()
```

```python
# Mediana

df_medidas["altura"].median()
```

```python
# Moda

df_medidas["altura"].mode()
```

### Medida de dispersão

```python
# variancia

df_medidas.idade.var()
```

```python
# desvio padrão

df_medidas.idade.std()
```

```python
# Coeficiente de Variação - idade

df_medidas.idade.std() / df_medidas.idade.mean() * 100
```

```python
# Coeficiente de Variação - altura

df_medidas.altura.std() / df_medidas.altura.mean() * 100
```

## Medidas de forma

```python
# Assimetria - idade

df_medidas.idade.skew()
```

```python
# Assimetria - Altura

df_medidas.altura.skew()
```

```python
# Curtose - idade

df_medidas.idade.kurtosis()
```

```python
# Obter medidas estatísticas

df_medidas.idade.describe()
```

### Correlação de variaveis

```python
# com base que a idade aumenta a altura também aumenta essa é a correlação
df_medidas.corr(method='pearson')
```

```python
# correlação entre duas variáveis
df_medidas.idade.corr(df_medidas.altura)
```

```python
# Correlação do DataFrame, segundo Sperman
df_medidas.corr(method='spearman')
```

### Criar Dataframe de Vendas

```python
dict_vendas = {
    'categoria': ['masculino', 'feminino', 'infantil', 'casa'],
    'valor': [4000, 6000, 250000, 540000],
    'quantidade': [3000, 5000, 1500, 2500]
}
```

```python
dict_vendas
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
df_vendas
```

```python
# Gráfico de Barras Vertical - Valor
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
# Gráfico de barras Vertical - Valor (ordenado)
df_vendas.sort_values('valor', ascending=False).plot.bar(x='categoria', y='valor')
```

```python
# Gráfico de Barras Horizontal - Quantidade
df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
# Gráfico de barras Horizontal - QUantifdade(Ordenado - Ranking)
df_vendas.sort_values('quantidade',ascending=True).plot.barh(x='categoria', y='quantidade')

#sort_values se não definir o ascending ele fica como true
```

```python
# Gráfico de Dispersão
df_medidas.plot.scatter(x='idade', y='altura')
```

```python
# Gráfico de BoxPlot (Diagrama de Caixa) - Idade
df_medidas.idade.plot.box()
```

```python
# Gráfico de BoxPlot - Altura
df_medidas.altura.plot.box()
```
